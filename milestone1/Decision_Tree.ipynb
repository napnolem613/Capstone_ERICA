```python
def get_groups(df):
    group=''
    if df['성별']==1:
        if df['연령(10대)']==1: group='남_10대'
        elif df['연령(10대)']==2: group='남_20대'
        elif df['연령(10대)']==3: group='남_30대'
        elif df['연령(10대)']==4: group='남_40대'
        elif df['연령(10대)']==5: group='남_50대'
        elif df['연령(10대)']==6: group='남_60대'
            
    elif df['성별']==2:
        if df['연령(10대)']==1: group='여_10대'
        elif df['연령(10대)']==2: group='여_20대'
        elif df['연령(10대)']==3: group='여_30대'
        elif df['연령(10대)']==4: group='여_40대'
        elif df['연령(10대)']==5: group='여_50대'
        elif df['연령(10대)']==6: group='여_60대'
    return group

answers['group']=(answers.loc[:,['성별','연령(10대)']]).apply(lambda x : get_groups(x),axis=1)
answers[['성별','연령(10대)','group']].head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
group_target = le.fit_transform(answers['group'])
print(group_target)
print(type(group_target))
print(group_target.shape)

# 변환된 label classes 확인
le.classes_
```

```python
target_names=le.classes_
print(target_names)
print(type(target_names))
print(target_names.shape)
```

```python
le.inverse_transform(group_target)
```

```python
# 원래 있던 성별과 나이는 새로 만든 group category에 포함되므로 제외하고, 지역과 직업은 category형 value가 아닌 연속적 값이고
# 또 추천 시스템에서 고려하지 않을 요소이므로 이를 제외하면 학습에 총 30개의 feature를 활용한다.

print(answers.loc[:,'광고 집중도 - 웹/앱 배너광고':'유아 아동용 화장품/바디용품 광고'].to_numpy())
print(answers.loc[:,'광고 집중도 - 웹/앱 배너광고':'유아 아동용 화장품/바디용품 광고'].to_numpy().shape)
print(type(answers.loc[:,'광고 집중도 - 웹/앱 배너광고':'유아 아동용 화장품/바디용품 광고'].to_numpy()))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

#하이퍼 파라미터 min_samples_splits=4, min_samples_leaf=4

dt_clf=DecisionTreeClassifier(min_samples_split=4, min_samples_leaf=4, random_state=156)
X_train,X_test,y_train,y_test=train_test_split(answers.loc[:,'광고 집중도 - 웹/앱 배너광고':'유아 아동용 화장품/바디용품 광고'].to_numpy(),group_target)

dt_clf.fit(X_train,y_train)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(dt_clf,out_file="tree.dot", class_names=target_names, \
                feature_names=features, impurity=True, filled=True)
```

```python
import graphviz

with open("tree.dot") as f:
    dot_graph=f.read()
graphviz.Source(dot_graph)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline

rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False  

# feature importance 분석

print("Feature importances:\n{0}".format(np.round(dt_clf.feature_importances_,3)))

for name,value in zip(features, dt_clf.feature_importances_):
    print('{0} : {1:.3f}'.format(name,value))
    
fig, ax = plt.subplots(figsize=(10, 8))
sns.barplot(x=dt_clf.feature_importances_,y=features,ax=ax)
plt.show()
```
